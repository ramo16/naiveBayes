```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv('F:\MachineLearning\FTSamples-PR\RSamples\diabetes.csv')

print(dataset.describe())

print(dataset.isnull().sum())

dataset['Glucose'] = dataset['Glucose'].replace('0', np.nan)
dataset['BloodPressure'] = dataset['BloodPressure'].replace('0', np.nan)
dataset['SkinThickness'] = dataset['SkinThickness'].replace('0', np.nan)
dataset['Insulin'] = dataset['Insulin'].replace('0', np.nan)
dataset['BMI'] = dataset['BMI'].replace('0', np.nan)
#dataset['DiabetesPedigreeFunction'] = 
dataset['DiabetesPedigreeFunction'].replace('0', np.nan)
dataset['Age'] = dataset['Age'].replace('0', np.nan)

print(dataset.head(10))
print(dataset.isnull().sum())

#Fill Missing Values
dataset['BMI'].fillna(dataset['BMI'].median(), inplace=True)
dataset['Glucose'].fillna(dataset['Glucose'].median(), inplace=True)
dataset['BloodPressure'].fillna(dataset['BloodPressure'].median(), inplace=True)
dataset['SkinThickness'].fillna(dataset['SkinThickness'].median(), inplace=True)
dataset['Insulin'].fillna(dataset['Insulin'].median(), inplace=True)

dataset.describe()
#Correlation Plot

corr = dataset[dataset.columns].corr()
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, cmap=cmap, annot = True)

#X = dataset.iloc[:, 0:8]
X_features = pd.DataFrame(data = dataset, columns = ["Glucose","BMI","Age"])
X_features.head(2)

y = dataset.iloc[:, 8]
#sns.heatmap(X.corr(), annot = True)
scaler = StandardScaler()
X_features = scaler.fit_transform(X_features)
print(X_features)

X_train, X_test, y_train, y_test = train_test_split(X_features, y, 
                                                    random_state=0, test_size=0.25)

# sc_X = StandardScaler()
# X_train = sc_X.fit_transform(X_train)
# X_test = sc_X.transform(X_test)

#Generic Models Accuracy
models = []

#models.append(("Logistic Regression:",LogisticRegression()))
models.append(("Naive Bayes:",GaussianNB()))
models.append(("K-Nearest Neighbour:",KNeighborsClassifier(n_neighbors=3)))
models.append(("Decision Tree:",DecisionTreeClassifier()))
models.append(("Support Vector Machine-linear:",SVC(kernel="linear",C=0.2)))
models.append(("Support Vector Machine-rbf:",SVC(kernel="rbf")))
models.append(("Random Forest:",RandomForestClassifier(n_estimators=5)))

results = []
names = []
for name,model in models:
    kfold = KFold(n_splits=5, random_state=3)
    cv_result = cross_val_score(model,X_train,y_train, cv = kfold,scoring = "accuracy")
    names.append(name)
    results.append(cv_result)
for i in range(len(names)):
    print(names[i],results[i].mean()*100)


classifier = GaussianNB()
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
print (cm)
print(accuracy_score(y_test, y_pred))
```
